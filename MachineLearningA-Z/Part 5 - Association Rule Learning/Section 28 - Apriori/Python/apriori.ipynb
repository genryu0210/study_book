```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, dataset.__len__()):
  transactions.append([str(dataset.values[i, j]) for j in range(1, dataset.columns.__len__())])
transactions
```

## モデルの訓練

```
from apyori import apriori
rules = apriori(transactions=transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
```

## 結果の可視化

### aprioriのデフォルトの結果の表示

```
results = list(rules)
```

```
results
```

### 情報を整理して表示するための関数

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][0])[0] for result in results]
    suppport    = [result[1] for  result in results]
    confidence  = [result[2][0][2] for  result in results]
    lift        = [result[2][0][3] for  result in results]
    return list(zip(lhs, rhs, suppport, confidence, lift))
resultsinDataFrame = pd.DataFrame(inspect(results), columns=['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

### 表による結果の表示

```
resultsinDataFrame
```
